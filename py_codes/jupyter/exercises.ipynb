```python
class Scoop(object):
    "Show the flavor of icecream scoop"
    def __init__(self, flavor):
        self.flavor = flavor

```

```python
s1 = Scoop('chocolate')
s2 = Scoop('vanilla')
s3 = Scoop('coffee')

print (s1.flavor)

for one_scoop in [s1, s2, s3]:
    print(one_scoop.flavor)
```
